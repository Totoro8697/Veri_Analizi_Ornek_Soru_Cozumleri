### - Burada bizden istenen sey belli. Hangileri hem is hem ev adresidir bulmamiz. 

### - Soyle bir durum var. Bunu yaparken karakter uzunlugundan yararlanmak dusunulebilir lakin bazi noktalarda kisaltmalar kullanildigindan bu cozum ise yaramayacaktir.

### - Bu yuzden her iki yerdeki adreslerin NO'sunu karsilastirmayi daha mantikli buldum.

### - Goruldugu uzere adresler tipatip ayni iken NO farkli olabiliyor

### - Burada soyle bir yol izliyoruz. Adres icerisinden NO'ya denk gelen integer degeri cikartiyoruz.

### - Ardindan bunu hem is adresi hem ev adresi olan yerlere uyguluyoruz. Cikarttigimiz integer deger uyusuyorsa bu adreslerin ayni yani hem ev hem de is adresi aynidir diyoruz. Aksi halde ev ve is adresleri farklidir durumuna ulasiyoruz.

```python
import numpy as np
import pandas as pd
```

```python
import re

ev_adres = ["Tugay Yolu Cad. No:60 Maltepe","İnkılap Mahallesi Küçüksu Caddesi No Ümraniye 6 ","Soğanlık Mah. Muş Sk. No 61 Kartal"]
is_adres = ["Tugay Yolu Caddesi No60 MALTEPE","İnkılap Mah. Küçüksu Cd. No 6 Ümraniye","Soğanlık Mah. Muş Sk. No 63 Kartal "]

no_values = []

for i,j in zip(ev_adres,is_adres):
    value_1 = re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", i)
    value_2 = re.findall("[-+]?[.]?[\d]+(?:,\d\d\d)*[\.]?\d*(?:[eE][-+]?\d+)?", j)
    
    if value_1 == value_2:
        print("Is ve Ev adresleri aynidir cunku NO uyusuyor\n","Ev Adresi No: ",value_1,"\n Is adresi NO: ",value_2)
        
    else:
        print("\nIs ve Ev Adresleri farklidir cunku kayitli NO uyusmuyor\n","Ev Adresi No: ",value_1,"\n Is adresi NO: ",value_2)
```
