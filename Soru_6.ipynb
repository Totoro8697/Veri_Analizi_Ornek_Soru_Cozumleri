### - Bu soruda adres bilgileri icerisinden TCNO ayirt etmemiz gerekiyor. Bunu yaparken ornekteki metinden bir df olusturup devam ediyoruz.

```python
import pandas as pd
import numpy as np

adresler = {"adres":["Tugay Yolu Cad. No:60 12537489122 Maltepe",
"İnkılap Mah. 88372637771 Küçüksu Cd. No 6 Ümraniye",
"11293888122 Soğanlık Mah. Muş Sk. No 63 Kartal",
"Bilezikçi sk 147 Şişli İstanbul 12229386"]}

adresler = pd.DataFrame(adresler)
adresler.head()
```

### - Ardindan sunu yapiyoruz. Adresleri bir listeye aliyoruz. Daha sonra yazdigimiz dongu icerisinde sirayla tum adreslere gidiyor ve her adres satirini bosluklardan itibaren parcaliyor onlari listenin icerisinde bir eleman haline geitiriyor. Ardindan her elemani isdigit yaparak yani sayi olup olmadigini sorgulyuor. Buradan sayi olanlari integer tipinde listemize ekliyoruz.

```python


strings = []
for word in adresler["adres"]:
    strings.append(word)

index = 0
numbers = []

for j in range(0,len(strings)):
    for word in strings[j].split():
        if word.isdigit():
            numbers.append(int(word))
              
            
```

```python
numbers
```

### - Simdi goruldugu uzere bu sayilar arasinda NO degerleri de gelmis bunlari elimine etmek icin gelen 4 karkater uzunlugundan daha fazla olanlari aliyoruz.

```python
tc_no = []

for j in numbers:
    if len(str(j)) > 4:
        tc_no.append(j)
        

        
adresler["Tc_No"] = np.array(tc_no)
```

```python
asd = []

index = 0
for i in tc_no:
    asd.append(adresler["adres"].values[index].replace(str(i),""))
    index += 1
```

### - Sonuclarin son hali bu sekilde elde ediliyor. Burada son satirda 11 haneden daha az 

```python
df_new = {"Adresler":asd,"Tc_No":adresler["Tc_No"]}
df_new = pd.DataFrame(df_new)
df_new.head()
```
